```python
import barcode
from barcode.writer import ImageWriter
from IPython.display import Image, display
```

```python
inputCode = input('Enter the code to generate barcode: ')
try:
    if len(inputCode) != 12:
        codeType = 'Code128'
    else:
        codeType = 'EAN13'
    print(codeType)
    my_code = barcode.get(codeType, inputCode, writer=ImageWriter())
    filename = input('Write a filename: ')
    my_code.save(filename)
    print(f"Code was successfully generated as '{filename}.png'")
except Exception as e:
    print(f"Error generating barcode {e}")

display(Image(filename=f'{filename}.png'))
```
